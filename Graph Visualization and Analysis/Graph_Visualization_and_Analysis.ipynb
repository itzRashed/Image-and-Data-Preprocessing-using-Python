```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from([1, 2, 3, 4, 5])

G.add_edges_from([(1, 1),(1, 2), (1, 3), (2, 3), (1, 4), (1, 5), (2, 5),(3, 4),(3, 5),(4, 5)])

nx.draw(G, with_labels=True, font_weight='bold')
plt.show()
```

```python
#Basic graph info
print("Nodes:",G.nodes())
print("Edges:",G.edges())
print("No of Nodes:",G.number_of_nodes())
print("No of Edges:",G.number_of_edges())
```

```python
#Check if the graph is connected
print(nx.is_connected(G))

#Shortest path
short_path=nx.shortest_path(G,source=1,target=5)
print("Shortest path from 1 to 5: ",short_path)
```

```python
#Degree of each node
degree=G.degree()
print("Degree of each node:",degree)

#For specific node
deg=G.degree(1)
print("Specific degree of a node:",deg)

deg_cen=nx.degree_centrality(G) #No of edges connected to 1
print("Degree of centrality",deg_cen)
```

**Centrality** = Degree of nodes/Total no of nodes - 1

**Clustering** = 2 X no of triangles connected to a node/Degree of  node X (degree of node - 1)

```python
clustering_coefficient=nx.clustering(G)
print("Clustering Coefficient:",clustering_coefficient)
```

```python
#Subgraph
sub_graph = G.subgraph([1,2,4])

nx.draw(sub_graph,with_labels=True,font_weight='bold')
plt.show()
```

```python
#To copy  a graph
G_copy = G.copy()

#To remove node
G_copy.remove_node(5)
nx.draw(G_copy,with_labels=True,font_weight='bold')
plt.show()
```

```python
#To remove edge
G1=G.copy()
G1.remove_edge(1,2)
nx.draw(G1,with_labels=True,font_weight='bold')
plt.show()
```

```python
#Find the component of a graph
components = list(nx.connected_components(G1))
print("Connected Components",components)

for i,component in enumerate(components):
  subgraph = G1.subgraph(component)
  plt.figure
  nx.draw(subgraph,with_labels=True,font_weight='bold')
  plt.show()
print("No of connected components:",i+1)


```
